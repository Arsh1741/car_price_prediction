```python
import pandas as pd
import numpy as np
df=pd.read_csv('car_data.csv')
df.info()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.drop(columns='unnamed:_0')
df.info()
```

```python
X=df.drop(labels=['price'], axis=1)
Y=df['price']
```

```python
X.head(1)
```

```python
Y.head(1)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest=train_test_split(X, Y, test_size=0.2, random_state=21)
```

```python
def cat_con(x):
    cat=[]
    con=[]
    for i in x.columns:
        if x[i].dtypes==object:
            cat.append(i)
        else:
            con.append(i)
    print('cat=',cat)
    print('con=',con)


cat_con(X)
```

```python
cat= ['name', 'company', 'fuel_type']
con= ['year', 'kms_driven']
```

```python
'''Xcon=X[con]
Xcat=X[cat]'''
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder, OrdinalEncoder
```

```python
ohe=OneHotEncoder()
ohe.fit(X[cat])
```

```python
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
col_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),cat), remainder='passthrough')
col_trans
```

```python
from sklearn.linear_model import LogisticRegression, LinearRegression
lr=LinearRegression()
```

```python
pipe = make_pipeline(col_trans, lr)
pipe
```

```python
pipe.fit(xtrain,ytrain)
```

```python
tr_pred=pipe.predict(xtrain)
ts_pred=pipe.predict(xtest)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(tr_pred, ytrain), r2_score(ts_pred, ytest)
```

```python
scores=[]
for i in range(1000):
    xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(col_trans,lr)
    pipe.fit(xtrain,ytrain)
    y_pred=pipe.predict(xtest)
    scores.append(r2_score(ytest,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
pipe.predict(pd.DataFrame(columns=xtest.columns,data=np.array(['Maruti Suzuki Swift','Maruti',2019,100,'Petrol']).reshape(1,5)))
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(col_trans,lr)
pipe.fit(xtrain,ytrain)
y_pred=pipe.predict(xtest)
r2_score(ytest,y_pred)
```
